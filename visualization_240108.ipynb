## matplotlib와 seaborn을 동시에 사용한 subplooting 기법
 - 다중차트를 하나의 도화지에 담도록 하겠다.

```python
fig, ax = plt.subplots(ncols=2)

plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2) # ==> row는 1개 columns은 2개로 그래프를 나눠주는 매개변수들이다. 매우중요!!!

ax[0].set_title("ax[0] titel") # ==> ax[0].set_title("ax[0] titel") [0]번째 인덱스에 해당하는 그래프에 이름을 만들어주는 메서드이다.
ax[0].plot([1,2,3,4])

ax[1].set_title("ax[1] titel")
sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'sex', ax=ax[1], data = tips)

fig.suptitle("fig.title") # ==> fig.title이라는 큰 제목을 만들어주는 메서드이다.
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2) # ==> matplotlib로 만드는 시각화 그래프 코드이다.!! 
ax[0,0].set_title("ax[0,0] titel")
ax[0,0].plot([1,2,3,4])

ax[0,1].set_title("ax[0,1] titel")
sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'sex', ax=ax[0,1], data = tips) # ==> seaborn으로 만드는 시각화 코드!!
# ==> sns.scatterplot()의 매개변수 중에 ax라는 매개변수 잘 찾아보기!!! 
# 여기서는 matplotlib그래프 ax[0,1] 위치에 seaborn의 그래프를 넣어주겠다. 즉 다중그래프를 그리겠다는 의미로 쓰인것 같다. ax=ax[0,1]이라는 매개변수를 써줌으로써!!!
# 종합하면 matplotlib그래프와 seaborn그래프를 연결시켜주는 매개변수이다.!!!!!!!!

fig.suptitle("fig.title")
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax[0,0].set_title("ax[0,0] titel")
ax[0,0].plot([1,2,3,4])

ax[1,1].set_title("ax[1,1] titel")
sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'sex', ax=ax[1,1], data = tips)

fig.suptitle("fig.title")
plt.tight_layout() # ==> 타이트하게 있던 그래프의 간격을 완화하게 해주는 메서드!!!
plt.show()
```

## 산점도 그래프
- x축과 y축 데이터 형태가 연속형 데이터가 와야 함
- 연속형 데이터의 예시 : 키, 몸무게 등 소수점이 존재하는 데이터

```python
fig, ax=plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
# fig, ax=plt.subplots(1, 2, figsize=(15, 5)) 매개변수중 nrows, ncols는 생략가능하다.!!!!

sns.regplot(x="total_bill", y="tip", data=tips, ax=ax[0], fit_reg=True)
ax[0].set_title('with linear regression line')

sns.regplot(x="total_bill", y="tip", data=tips, ax=ax[1], fit_reg=False)
ax[1].set_title('without linear regression line')

plt.show()
```

## 라이브러리 가져오기
- 시각화 중급 레벨

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator, FuncFormatter) # 축설정할 때 값
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## 객체지향 선언

```python
# y축 설정할 때, 값을 소수점 2자리까지 표현
def major_formatter(x, pos):
    return "%.2f$" % x
formatter = FuncFormatter(major_formatter)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16,5))

# 이상적인 막대 그래프
ax0 = sns.barplot(x = "day", y = 'total_bill', data = tips,
                  errorbar=None, color='lightgray', zorder=2, alpha=0.85, ax=ax[0])
                                              # ==> zorder=2는 그래프를 그렸을 때 grid선을 그래프 뒤로 안보이게 해준다.

group_mean = tips.groupby(['day'])['total_bill'].agg('mean')
h_day = group_mean.sort_values(ascending=False).index[0] # 해당되는 결과값의 MAX값의 인덱스 요일을 하나 가져옴.
h_mean = np.round(group_mean.sort_values(ascending=False)[0], 2) # 해당되는 결과값의 MAX값을 하나 가져옴.
for p in ax0.patches:
  fontweight = "normal"
  color = "k"
  height = np.round(p.get_height(), 2)
  if h_mean == height:
    fontweight="bold"
    color="darkred"
    p.set_facecolor(color)
    p.set_edgecolor("black")
  ax0.text(p.get_x() + p.get_width()/2., height+1, height, ha = 'center',
           size=12,
           fontweight=fontweight, color=color)

# y축 범위 지정 옵션
ax0.set_ylim(-3, 30)
ax0.set_title("Ideal Bar Graph", size = 16)

# 가로세로 축, 일부 수정 및 변경
ax0.spines['top'].set_visible(False)
ax0.spines['left'].set_position(("outward", 20))
ax0.spines['left'].set_visible(False)
ax0.spines['right'].set_visible(False)

# y축 값 변경
ax0.yaxis.set_major_locator(MultipleLocator(10))
ax0.yaxis.set_major_formatter(formatter)
ax0.yaxis.set_minor_locator(MultipleLocator(5))

# y축 라벨 변경
ax0.set_ylabel("Avg. Total Bill($)", fontsize=14)

# y축 Grid 라인 수정
ax0.grid(axis="y", which="major", color="lightgray")
ax0.grid(axis="y", which="minor", ls=":")

# x축의 값을 변경
ax0.set_xlabel("Weekday", fontsize=14)
for xtick in ax0.get_xticklabels():
  print(xtick)
  if xtick.get_text() == h_day:
    xtick.set_color("darkred")
    xtick.set_fontweight("demibold")
ax0.set_xticklabels(['Thursday', 'Friday', 'Saturday', 'Sunday'], size=12)

# 그냥 일반 막대 그래프
ax1 = sns.barplot(x = "day", y = 'total_bill', data = tips, errorbar=None, alpha=0.85, ax=ax[1])
for p in ax1.patches:
  height = np.round(p.get_height(), 2)
  ax1.text(p.get_x() + p.get_width()/2., height+1, height, ha = 'center', size=12)
ax1.set_ylim(-3, 30)
ax1.set_title("Just Bar Graph")

plt.show()
```
