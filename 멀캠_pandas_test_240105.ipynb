## (2) 데이터 불러오기
- tips 데이터를 불러와서 tips라고 저장한다.

```python
tips=sns.load_dataset('tips')
```

## (3) 특정 행 추출
- loc을 사용하여 tips 데이터셋의 처음 3행의 데이터를 추출하세요.

```python
# loc와 iloc 차이점
 #- loc에 l은 무엇을 의미할까요? l은 label 기반이다.
 #- iloc에 i는 무엇을 의미할까요? i는 index 기반이다.
```

```python
tips.loc[0:3, :] # ==> loc는 label기반이기에 [0:3]까지 잡아도 label 3까지 나오게 된다.
```

```python
tips.iloc[0:3, :] # ==> iloc는 index기반이기에 [0:3]까지 범위를 잡으면 index기반에 0번부터 2번까지만 나오게 된다.
```

```python
tips.loc[:3, :]
```

## (4) 특정 열 추출
- loc을 사용하여 모든 레코드에 대한 'total_bill'과 'tip' 열을 추출하세요.

```python
tips.loc[ :, ['total_bill', 'tip']]
```

```python
tips.loc[:, ['total_bill']] # ==> 열의 조건식에서 매개변수가 한개라도 []리스트로 감싸는 것 잊지말기!!
                            # ==> 열의 조건식에서 매개변수가 2개 이상이면 무조건 []리스트로 감싸서 써야한다.
```

## (5) 특정 행과 열의 조합
- loc을 사용하여 처음 5행에 대해 'total_bill', 'tip', 'day' 열을 추출하세요.

```python
tips.loc[:5, ['total_bill', 'tip', 'day']]
```

## (6) 조건에 따른 행 추출
- 'day'가 'Sunday'인 모든 행을 추출하고 'total_bill'과 'tip' 열만 표시하세요.

```python
tips.loc[tips['day']=='Sun', ['total_bill', 'tip']]
```

## (7) 다중 조건에 따른 행 추출
- loc을 사용하여 'time'이 'Dinner'이고 'tip'이 5달러 이상인 모든 행을 선택하고 'total_bill', 'sex', 'smoker' 열만 표시하세요.

```python
result=tips.loc[(tips['time']=='Dinner') & (tips['tip']>=5), ['total_bill', 'sex', 'smoker', 'tip']]
result # ==> 행의 숫자를 보면 즉 index의 번호를 보면 숫자가 정렬이 되어있지 않다.
```

```python
result.reset_index(drop=True) # ==> reset_index(drop=True)는 행의 번호를 즉 index의 번호를 깔끔하게 정리해주는 메서드이다.
```
