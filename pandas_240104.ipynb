```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# (1)total_bill의 중간값 이하인 것만 조회
midium=tips['total_bill'].median()
tips.loc[tips['total_bill']<=midium,['total_bill']]
```

```python
# (2)day가 Sun인 데이터만 조회하기
tips.loc[tips['day']=='Sun', ['day']]
```

```python
# (3)tip의 평균보다 크면서, time이 Dinner인 데이터만 조회
avg_tip=tips['tip'].mean()
avg_tip

tips.loc[
    tips['tip'] > avg_tip & tips['time'] == 'Dinner', :
]
```

```python
avg_tip=tips['tip'].mean()
avg_tip

tips.loc[
    (tips['tip'] > avg_tip) & (tips['time'] == 'Dinner'), :
] 
  # ==> 다중조건문을 할 때에는 (조건1) & (조건2) 이렇게 써주어야 한다. ()로 꼭 감싸주어야 한다. 
  #     위에 예시를 봐도 다중조건일 때 각자의 조건에 ()을 안붙히면 error가 나온다.
  # & AND조건, | OR조건 이다.
```

```python
tips.loc[10, :] # 1.행을 추출하는 방법이다.
```

```python
tips.loc[10:10, :] # 2.행을 추출하는 방법이다.
```
